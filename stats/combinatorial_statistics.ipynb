# Examples Exercises From Trading

* Compute the number of combinations between the following currencies: USD, EUR,CAD. (We consider that EUR/USD = USD/EUR).


```python
import numpy as np

def combination(n, k, repetition = False):
    """
    A function that calculates the number of combinations given n,k. 
    The function accounts for both repeating and non-repeating values
    """
    if repetition:
        # If repetiton is allowed
        nCk = np.math.factorial(n + k - 1)/(np.math.factorial(n-1)* np.math.factorial(k))
    else:
        # If repetiton is not allowed
        nCk = np.math.factorial(n)/(np.math.factorial(n-k)* np.math.factorial(k))
    return nCk
```

```python
# Pick a combination of 2 from a group of 3 . Therefore n = 3, k = 2. 
# Since the currency ratios have to be different to be n new valid combination,
# repetition is not allowed.
combination(3,2, repetition = False)
```

* Compute the number of combinations between the following currencies: USD, EUR,CAD. (We consider that EUR/USD != USD/EUR)

```python
# Repetition is allowed since two of the 
# same values give different combinations
combination(3,2, repetition = True)
```
